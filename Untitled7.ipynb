```python
import pandas as pd
import matplotlib.pyplot as plt

def main():
    # 1. Importation des données
    print("\n1. IMPORTATION DES DONNÉES")
    try:
        df = pd.read_csv("StudentsPerformance.csv")
        print("Fichier CSV chargé avec succès!")
        print(f"Nombre de lignes/colonnes : {df.shape}")
    except FileNotFoundError:
        print("Erreur: Fichier non trouvé. Veuillez vérifier le chemin.")
        return
    
    # 2. Exploration des données
    print("\n2. EXPLORATION DES DONNÉES")
    print("\nA. Premières lignes du dataset:")
    print(df.head())
    
    print("\nB. Informations générales:")
    print(df.info())
    
    print("\nC. Statistiques descriptives:")
    print(df.describe(include='all'))
    
    print("\nD. Vérification des valeurs manquantes:")
    print(df.isnull().sum())
    
    # 3. Manipulation des données
    print("\n3. MANIPULATION DES DONNÉES")
    
    # Sélection de colonnes
    scores = df[['math score', 'reading score', 'writing score']]
    print("\nA. Colonnes des scores:")
    print(scores.head())
    
    # Filtrage des données
    female_students = df[df['gender'] == 'female']
    print("\nB. Nombre d'étudiantes:", len(female_students))
    
    # Création d'une nouvelle colonne (moyenne des scores)
    df['average score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1).round(2)
    print("\nC. DataFrame avec nouvelle colonne 'average score':")
    print(df.head())
    
    # Tri des données
    top_students = df.sort_values('average score', ascending=False).head(10)
    print("\nD. Top 10 des étudiants par moyenne:")
    print(top_students[['gender', 'race/ethnicity', 'parental level of education', 'average score']])
    
    # 4. Analyse statistique
    print("\n4. ANALYSE STATISTIQUE")
    
    # Statistiques descriptives par groupe
    print("\nA. Statistiques par genre:")
    print(df.groupby('gender')['average score'].describe())
    
    print("\nB. Moyenne des scores par groupe ethnique:")
    print(df.groupby('race/ethnicity')[['math score', 'reading score', 'writing score']].mean().round(2))
    
    print("\nC. Impact du cours de préparation:")
    prep_course = df.groupby('test preparation course')['average score'].mean()
    print(prep_course)
    improvement = (prep_course['completed'] - prep_course['none']) / prep_course['none'] * 100
    print(f"\nAmélioration moyenne avec préparation: {improvement:.2f}%")
    
    # 5. Visualisation des données
    print("\n5. VISUALISATION DES DONNÉES")
    
    # Configuration des graphiques
    plt.style.use('ggplot')
    plt.figure(figsize=(15, 10))
    
    # Histogramme des moyennes
    plt.subplot(2, 2, 1)
    df['average score'].plot.hist(bins=20, edgecolor='black')
    plt.title('Distribution des moyennes des étudiants')
    plt.xlabel('Moyenne')
    plt.ylabel('Nombre d\'étudiants')
    
    # Boxplot par genre
    plt.subplot(2, 2, 2)
    df.boxplot(column=['math score', 'reading score', 'writing score'], by='gender')
    plt.title('Scores par genre')
    plt.suptitle('')
    plt.ylabel('Score')
    
    # Moyenne par groupe ethnique
    plt.subplot(2, 2, 3)
    ethnic_avg = df.groupby('race/ethnicity')['average score'].mean()
    ethnic_avg.plot(kind='bar', color='teal')
    plt.title('Moyenne par groupe ethnique')
    plt.ylabel('Moyenne')
    plt.xticks(rotation=45)
    
    # Impact du cours de préparation
    plt.subplot(2, 2, 4)
    prep_course.plot(kind='bar', color=['red', 'green'])
    plt.title('Impact du cours de préparation')
    plt.ylabel('Moyenne')
    plt.xticks(rotation=0)
    
    plt.tight_layout()
    plt.savefig('student_performance_analysis.png')
    print("\nGraphiques sauvegardés dans 'student_performance_analysis.png'")
    
    # 6. Analyse supplémentaire
    print("\n6. ANALYSE COMPLÉMENTAIRE")
    
    # Relation entre niveau d'éducation des parents et performance
    parent_edu = df.groupby('parental level of education')['average score'].mean().sort_values()
    print("\nPerformance moyenne par niveau d'éducation des parents:")
    print(parent_edu.round(2))
    
    # Impact du type de déjeuner
    lunch_impact = df.groupby('lunch')['average score'].mean()
    print("\nImpact du type de déjeuner:")
    print(lunch_impact)
    difference = lunch_impact['standard'] - lunch_impact['free/reduced']
    print(f"\nDifférence entre standard et free/reduced: {difference:.2f} points")
    
    # Sauvegarde des résultats
    with open('analysis_results.txt', 'w') as f:
        f.write("RÉSULTATS DE L'ANALYSE\n")
        f.write("======================\n\n")
        f.write("1. Statistiques globales:\n")
        f.write(df.describe().to_string())
        f.write("\n\n2. Performance par genre:\n")
        f.write(df.groupby('gender')['average score'].describe().to_string())
        f.write("\n\n3. Impact du cours de préparation:\n")
        f.write(prep_course.to_string())
        f.write(f"\nAmélioration: {improvement:.2f}%")
        f.write("\n\n4. Performance par éducation des parents:\n")
        f.write(parent_edu.to_string())
    
    print("\nAnalyse terminée. Résultats sauvegardés dans 'analysis_results.txt'")

if __name__ == "__main__":
    main()
```
